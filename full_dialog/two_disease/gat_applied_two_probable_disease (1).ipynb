```python
class MyOwnDataset_complete(InMemoryDataset):
    def __init__(self, root, transform=None, pre_transform=None, pre_filter=None):
        super().__init__(root, transform, pre_transform, pre_filter)
        self.data, self.slices = torch.load(self.processed_paths[0])
        
    @property
    def processed_file_names(self):
        return ['data_complete_1367.pt']
    
    def process(self):
#         # Read data into huge `Data` list.
        data_list1 = data_list

        if self.pre_filter is not None:
            data_list1 = [data for data in data_list1 if self.pre_filter(data)]

        if self.pre_transform is not None:
            data_list1 = [self.pre_transform(data) for data in data_list1]

        data, slices = self.collate(data_list1)
        torch.save((data, slices), self.processed_paths[0])
```

```python
class GAT(torch.nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, args):
        super(GAT,self).__init__()
        
        self.conv1 = GATConv(input_dim, hidden_dim, heads=args['heads'])
        self.conv2 = GATConv(hidden_dim*args['heads'], hidden_dim, heads = args['heads'])
        

#         self.conv3 = GATConv(hidden_dim*args['heads'], hidden_dim, heads = args['heads'])
#         self.bn1 = BatchNorm1d(hidden_dim*args['heads'])
#         self.bn2 = BatchNorm1d(hidden_dim*args['heads'])
        
        self.lin = torch.nn.Linear(hidden_dim*args['heads'], output_dim)
#         input = 10,768
#         input_dim =768
#         self.conv1 = o/p 10X(hidden_dim*heads)
#         self.conv2 = o/p 10X(hidden_dim*heads)
    def forward(self, data, batch):
        x, edge_index = data.x, data.edge_index
        
        x=self.conv1(x, edge_index)
        x=F.dropout(F.relu(x), p=args['dropout'], training=self.training)
#         x = self.bn1(x)
#         x = F.relu(x)
        
        x=self.conv2(x, edge_index)
#         x = self.bn2(x)
#         x = F.relu(x)
#         x=F.dropout(F.relu(x), p=args['dropout'], training=self.training)
        
#         x=self.conv3(x, edge_index)
        
        
        
        x=global_mean_pool(x, batch)
#         x=global_add_pool(x, batch)
#         x=global_max_pool(x, batch)
        
        
#         x - 10X(hidden_dim*heads)
#         after global mean pool
#         batch - 8
#         o/p 8Xhidden_dim
        x=F.dropout(F.relu(x), p=args['dropout'], training=self.training)
        
        x = self.lin(x)
        
#         x = F.sigmoid(x)
        
#         x = x.view(-1)
#         x = torch.nn.Sigmoid(x)
        return x
        
```

```python
# optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
# optimizer = torch.optim.Adam(model.parameters(), lr=0.003)
# optimizer = torch.optim.Adam(model.parameters(), lr=0.0001)
# optimizer = torch.optim.Adam(model.parameters(), lr=0.00001)
criterion = torch.nn.CrossEntropyLoss()
```

```python
def train(loader):
    model.train()
    
    for data in loader:
        optimizer.zero_grad()
#         print(data.batch)
        batch = data.batch
        data = data.to(device)
        batch = batch.to(device)
        out = model(data, batch)
        # out = model(data, data.batch)
#         print('Prediction : ', out)
#         print('Prediction shape : ', out.shape)
#         print('Ground Truth Shape : ',data.y.shape)
#         print('Ground Truth : ',data.y)
#         print('Ground Truth data type : ',data.y.dtype)
#         print('Output type',out.dtype)
        z = data.y
        ground_truth = z.long()
#         print('Ground Truth data type : ',ground_truth.dtype)
#         print('Ground Truth shape : ',ground_truth.shape)
        
        loss = criterion(out, ground_truth)
        
#         outputs = F.log_softmax(out, dim=1)
#         print('After log softmax shape : ', outputs.shape)
#         loss = criterion(outputs, ground_truth)
        print('loss : ',loss)
        loss.backward()
        optimizer.step()
```

```python
def test(loader):
    model.eval()
    
    y_true = []
    y_pred = []

    correct = 0
    for data in loader:
      with torch.no_grad():
          batch = data.batch
          data = data.to(device)
          batch = batch.to(device)
          out = model(data, batch)
          # out = model(data, data.batch)
  #         print('Output embedding : ',out.shape)
          pred = out.argmax(dim=1)
  #         print('Prediction : ',pred)
  #         print('Ground Truth : ', data.y)
  #         print()
          correct += int((pred == data.y).sum())

          true_label = data.y

          y_true.extend(true_label.tolist())
          y_pred.extend(pred.tolist())


    return correct/len(loader.dataset), y_true, y_pred    
```

```python
def train_and_validation(model, train_loader, valid_loader):
  best_acc = 0.0
  

  for epoch in range(25):

    print("\n=============Epoch : ", epoch)
    train( train_loader)
    valid_acc, valid_pred, valid_gold = test(valid_loader)

    print("Length of predictions : ", len(valid_pred))
    print("Length of gold : ", len(valid_gold))
    # print("Valid loss : ", valid_loss)
    
    print("\nAccuracy\n", accuracy_score(valid_gold, valid_pred))
    print("\nF1 score\n", f1_score(valid_gold, valid_pred, average = 'macro'))
    print("\nJaccard similarity\n", jaccard_score(valid_gold, valid_pred, average = 'macro')) 

    curr_acc = accuracy_score(valid_gold, valid_pred)

    print('Curr acc : ', curr_acc)
    print('Best acc : ', best_acc)

    if(curr_acc > best_acc):
      best_acc = curr_acc

      
      torch.save(model.state_dict(), '/content/drive/MyDrive/Colab Notebooks/pakdd_baseline/saved_model/models/knowledge_only/two_disease/best_model_epoch_'+str(epoch)+'_best_acc_'+str(int(best_acc*100))+'_.pt')

      print("model saved\n")
```
